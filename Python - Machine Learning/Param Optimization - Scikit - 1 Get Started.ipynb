```python
import warnings
import skopt
import skopt.plots as skopt_plots

warnings.filterwarnings('ignore')
```

```python
def func(params): 
    A, B = params
    calc = [
        [10, 20, 30], 
        [40, 50, 60], 
        [70, 80, 90], 
    ]
    return calc[A][B]


space = [
    skopt.space.Categorical([0,1], name='A'),
    skopt.space.Categorical([0,1], name='B'),
]

res = skopt.gp_minimize(func, space)
res
```
