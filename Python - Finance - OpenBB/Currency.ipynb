```python
import pandas as pd
from openbb import obb
```

```python
def obb_currency_historical(SYMBOLS):
    Pipe = []
    for i, SYMBOL in enumerate(SYMBOLS): 
        try: 
            print(i, SYMBOL)
            X = obb.currency.price.historical(SYMBOL).results
            Pipe.append(pd.DataFrame([{ 'Symbol':SYMBOL, 'Date':x.date, 'Volume':x.volume, 'Price':x.close } for x in X]))
        except Exception as E: 
            print(i, SYMBOL, E)
    pass
    return pd.concat(Pipe, ignore_index=1) 

obb_currency_historical(['EUR','GBP'])
```
