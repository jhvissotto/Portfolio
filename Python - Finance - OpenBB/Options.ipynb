```python
import pandas as pd
from openbb import obb
```

```python
def obb_options_chains(TICKERS):
    Pipe = []
    for i, TICKER in enumerate(TICKERS): 
        try: 
            print(i, TICKER)
            X = obb.derivatives.options.chains(TICKER).results
            Pipe.append(pd.DataFrame({ 'Ticker':TICKER, 'Price':X.underlying_price, 'Option':X.option_type, 'Expiry':X.expiration, 'Strike':X.strike, 'Volume':X.volume, 'OI':X.open_interest, 'Ask':X.ask, 'Bid':X.bid }))
        except Exception as E: 
            print(i, TICKER, E)
    pass
    return pd.concat(Pipe, ignore_index=1) 

OChain = obb_options_chains(['MSFT','GOOG','META'])
OChain.head()
```
