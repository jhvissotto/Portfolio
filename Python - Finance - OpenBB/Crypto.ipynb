```python
import pandas as pd
from openbb import obb
```

```python
def obb_crypto_historical(PAIRS):
    Pipe = []
    for i, PAIR in enumerate(PAIRS): 
        try: 
            print(i, PAIR)
            X = obb.crypto.price.historical(PAIR).results
            Pipe.append(pd.DataFrame([{ 'Pair':PAIR, 'Date':x.date, 'Volume':x.volume, 'Price':x.close } for x in X]))
        except Exception as E: 
            print(i, PAIR, E)
    pass
    return pd.concat(Pipe, ignore_index=1) 

obb_crypto_historical(['BTC-USD', 'ETH-USD'])
```
