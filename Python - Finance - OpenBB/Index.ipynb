```python
import pandas as pd
from openbb import obb
```

```python
def obb_index_available():
    X = obb.index.available().results
    return pd.DataFrame([{ 'Symbol':x.symbol, 'Code':x.code, 'Name':x.name } for x in X])

obb_index_available()
```

```python
def obb_index_historical(TICKERS):
    Pipe = []
    for i, TICKER in enumerate(TICKERS): 
        try: 
            print(i, TICKER)
            X = obb.index.price.historical(TICKER).results
            Pipe.append(pd.DataFrame([{ 'Symbol':TICKER, 'Date':x.date, 'Volume':x.volume, 'Price':x.close } for x in X]))
        except Exception as E: 
            print(i, TICKER, E)
    pass
    return pd.concat(Pipe, ignore_index=1) 

obb_index_historical(['^GSPC','^SPX'])
```
