```python
import pandas as pd
from openbb import obb
```

```python
def obb_futures_curve(TICKERS):
    Pipe = []
    for i, TICKER in enumerate(TICKERS): 
        try: 
            print(i, TICKER)
            X = obb.derivatives.futures.curve(TICKER).results
            Pipe.append(pd.DataFrame([{ 'Ticker':TICKER, 'Expiry':x.expiration, 'Future':x.price } for x in X]))
        except Exception as E: 
            print(i, TICKER, E)
    pass
    return pd.concat(Pipe, ignore_index=1) 

obb_futures_curve(['NG'])
```

```python
def obb_futures_historical(TICKERS):
    Pipe = []
    for i, TICKER in enumerate(TICKERS): 
        try: 
            print(i, TICKER)
            X = obb.derivatives.futures.historical(TICKER).results
            Pipe.append(pd.DataFrame([{ 'Ticker':TICKER, 'Date':x.date, 'Volume':x.volume, 'Price':x.close } for x in X]))
        except Exception as E: 
            print(i, TICKER, E)
    pass
    return pd.concat(Pipe, ignore_index=1) 

obb_futures_historical(['NG'])
```
