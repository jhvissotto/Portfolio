```python
import matplotlib.pyplot as plt
```

# Line Chart

```python
x = [ 1,  2,  3,  4,  5 ]
y = [10, 20, 40, 80, 160]

plt.title('Line Chart')
plt.plot(x, y, label='y=f(x)')
plt.yscale('log', base=2)
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.tight_layout()
plt.show()
```

# Scatter Chart

```python
x = [ 1,  2,  3,  4,  5 ]
y = [10, 20, 40, 80, 160]

plt.title('Scatter Chart')
plt.scatter(x, y, label='y=f(x)')
plt.yscale('log', base=2)
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.tight_layout()
plt.show()
```

# Bar Chart

```python
x = [ 1,  2,  3,  4,  5 ]
y = [10, 20, 40, 80, 160]

plt.title('Bar Chart')
plt.bar(x, y, label='y=f(x)')
plt.yscale('log', base=2)
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.tight_layout()
plt.show()
```
