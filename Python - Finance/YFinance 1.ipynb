```python
import yfinance as yf
```

```python
TICKER = 'MSFT'
```

# Historical Data

```python
yf.download(TICKER)
```

# Quote

```python
info = yf.Ticker(TICKER).info

info['currentPrice'], info['previousClose'], info['ask'], info['bid']
```

# Options

```python
expirations = yf.Ticker(TICKER).options
expirations
```

```python
yf.Ticker(TICKER).option_chain(expirations[0]).calls.head()
```
