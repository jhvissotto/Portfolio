```python
import requests as req
import pandas as pd
```

```python
def SLICKCHARTS_CONSTITUENTS(indice):
    return pd.read_html(req.get(f'https://www.slickcharts.com/{indice}', headers={ 'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36', 'X-Requested-With':'XMLHttpRequest' }).content)[0]

US_30   = SLICKCHARTS_CONSTITUENTS('dowjones')
US_100  = SLICKCHARTS_CONSTITUENTS('nasdaq100')
US_500  = SLICKCHARTS_CONSTITUENTS('sp500')
```
